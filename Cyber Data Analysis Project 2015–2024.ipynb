```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Konfigurimi i stilit të grafikëve
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Leximi i të dhënave
print("=" * 80)
print("ANALIZA E TË DHËNAVE TË SIGURISË KIBERNETIKE 2015-2024")
print("=" * 80)
print("\n1. LEXIMI I TË DHËNAVE...")

df = pd.read_csv('/content/Sample Data/Global_Cybersecurity_Threats_2015-2024.csv')

# Informacione të përgjithshme
print(f"\nNumri total i incidenteve: {len(df)}")
print(f"Periudha kohore: {df['Year'].min()} - {df['Year'].max()}")
print(f"\nKolonat e dataset-it:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")

# Statistika përshkruese
print("\n" + "=" * 80)
print("2. STATISTIKA PËRSHKRUESE")
print("=" * 80)
print("\nStatistika për variablat numerike:")
print(df.describe())

# Kontrolli për vlera të munguara
print("\n" + "=" * 80)
print("3. KONTROLLI I VLERAVE TË MUNGUARA")
print("=" * 80)
missing_values = df.isnull().sum()
print("\nVlera të munguara për secilin kolonë:")
print(missing_values)
print(f"\nPerqindja totale e vlerave të munguara: {(df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100:.2f}%")

# Analiza e shpërndarjes së sulmeve sipas viteve
print("\n" + "=" * 80)
print("4. ANALIZA E TENDENCAVE KOHORE")
print("=" * 80)
attacks_by_year = df['Year'].value_counts().sort_index()
print("\nNumri i sulmeve sipas viteve:")
print(attacks_by_year)

# Analiza e llojeve të sulmeve
print("\n" + "=" * 80)
print("5. ANALIZA E LLOJEVE TË SULMEVE")
print("=" * 80)
attack_types = df['Attack Type'].value_counts()
print("\nLloje të sulmeve dhe frekuenca e tyre:")
print(attack_types)
print(f"\nPerqindja e secilit lloj sulmi:")
for attack, count in attack_types.items():
    print(f"  {attack}: {(count/len(df)*100):.2f}%")

# Analiza e shteteve më të prekura
print("\n" + "=" * 80)
print("6. ANALIZA GJEOGRAFIKE")
print("=" * 80)
countries = df['Country'].value_counts().head(10)
print("\nTop 10 shtetet më të prekura:")
print(countries)

# Analiza e industrive të synuara
print("\n" + "=" * 80)
print("7. ANALIZA E INDUSTRIVE TË SYNUARA")
print("=" * 80)
industries = df['Target Industry'].value_counts()
print("\nIndustritë e synuara dhe frekuenca:")
print(industries)

# Analiza e humbjeve financiare
print("\n" + "=" * 80)
print("8. ANALIZA E HUMBJEVE FINANCIARE")
print("=" * 80)
print(f"\nHumbja financiare totale: ${df['Financial Loss Million Dollar'].sum():.2f} milionë")
print(f"Humbja mesatare për incident: ${df['Financial Loss Million Dollar'].mean():.2f} milionë")
print(f"Humbja mediane: ${df['Financial Loss Million Dollar'].median():.2f} milionë")
print(f"Humbja maksimale: ${df['Financial Loss Million Dollar'].max():.2f} milionë")
print(f"Humbja minimale: ${df['Financial Loss Million Dollar'].min():.2f} milionë")

# Analiza e përdoruesve të prekur
print("\n" + "=" * 80)
print("9. ANALIZA E PËRDORUESVE TË PREKUR")
print("=" * 80)
print(f"\nNumri total i përdoruesve të prekur: {df['Number of Affected Users'].sum():,}")
print(f"Mesatarja e përdoruesve të prekur për incident: {df['Number of Affected Users'].mean():.0f}")
print(f"Mediana: {df['Number of Affected Users'].median():.0f}")

# Analiza e burimeve të sulmeve
print("\n" + "=" * 80)
print("10. ANALIZA E BURIMEVE TË SULMEVE")
print("=" * 80)
attack_sources = df['Attack Source'].value_counts()
print("\nBurimet e sulmeve:")
print(attack_sources)

# Analiza e dobësive të sigurisë
print("\n" + "=" * 80)
print("11. ANALIZA E DOBËSIVE TË SIGURISË")
print("=" * 80)
vulnerabilities = df['Security Vulnerability Type'].value_counts()
print("\nLloje të dobësive të sigurisë:")
print(vulnerabilities)

# Analiza e mekanizmave të mbrojtjes
print("\n" + "=" * 80)
print("12. ANALIZA E MEKANIZMAVE TË MBROJTJES")
print("=" * 80)
defense_mechanisms = df['Defense Mechanism Used'].value_counts()
print("\nMekanizma të mbrojtjes të përdorur:")
print(defense_mechanisms)

# Analiza e kohës së zgjidhjes
print("\n" + "=" * 80)
print("13. ANALIZA E KOHËS SË ZGJIDHJES SË INCIDENTEVE")
print("=" * 80)
print(f"\nKoha mesatare e zgjidhjes: {df['Incident Hour'].mean():.2f} orë")
print(f"Koha mediane e zgjidhjes: {df['Incident Hour'].median():.2f} orë")
print(f"Koha maksimale e zgjidhjes: {df['Incident Hour'].max():.2f} orë")
print(f"Koha minimale e zgjidhjes: {df['Incident Hour'].min():.2f} orë")

# Analiza e korrelacioneve
print("\n" + "=" * 80)
print("14. ANALIZA E KORRELACIONEVE")
print("=" * 80)
numeric_cols = ['Year', 'Financial Loss Million Dollar', 'Number of Affected Users',
                'Incident Hour']
correlation_matrix = df[numeric_cols].corr()
print("\nMatrica e korrelacioneve:")
print(correlation_matrix)

# Analiza e humbjeve financiare sipas llojit të sulmit
print("\n" + "=" * 80)
print("15. HUMBJET FINANCIARE SIPAS LLOJIT TË SULMIT")
print("=" * 80)
financial_by_attack = df.groupby('Attack Type')['Financial Loss Million Dollar'].agg(['sum', 'mean', 'count'])
financial_by_attack = financial_by_attack.sort_values('sum', ascending=False)
print("\nHumbjet financiare sipas llojit të sulmit:")
print(financial_by_attack)

# Analiza e humbjeve financiare sipas industrisë
print("\n" + "=" * 80)
print("16. HUMBJET FINANCIARE SIPAS INDUSTRISË")
print("=" * 80)
financial_by_industry = df.groupby('Target Industry')['Financial Loss Million Dollar'].agg(['sum', 'mean', 'count'])
financial_by_industry = financial_by_industry.sort_values('sum', ascending=False)
print("\nHumbjet financiare sipas industrisë:")
print(financial_by_industry)

# Analiza e tendencave kohore për humbjet financiare
print("\n" + "=" * 80)
print("17. TENDENCA KOHORE E HUMBJEVE FINANCIARE")
print("=" * 80)
financial_by_year = df.groupby('Year')['Financial Loss Million Dollar'].agg(['sum', 'mean', 'count'])
print("\nHumbjet financiare sipas viteve:")
print(financial_by_year)

# Ruajtja e rezultateve në një skedar
print("\n" + "=" * 80)
print("18. RUAJTJA E REZULTATEVE")
print("=" * 80)

# Krijimi i një përmbledhje
summary_data = {
    'Metrika': [
        'Numri total i incidenteve',
        'Periudha kohore',
        'Humbja financiare totale (milionë $)',
        'Humbja mesatare për incident (milionë $)',
        'Përdorues të prekur (total)',
        'Koha mesatare e zgjidhjes (orë)',
        'Lloji më i shpeshtë i sulmit',
        'Industria më e synuar',
        'Burimi kryesor i sulmit',
        'Dobësia më e shpeshtë e sigurisë'
    ],
    'Vlera': [
        len(df),
        f"{df['Year'].min()}-{df['Year'].max()}",
        f"{df['Financial Loss Million Dollar'].sum():.2f}",
        f"{df['Financial Loss Million Dollar'].mean():.2f}",
        f"{df['Number of Affected Users'].sum():,}",
        f"{df['Incident Hour'].mean():.2f}",
        attack_types.index[0],
        industries.index[0],
        attack_sources.index[0],
        vulnerabilities.index[0]
    ]
}

summary_df = pd.DataFrame(summary_data)
summary_df.to_csv('/content/Sample Data/analysis_summary.csv', index=False)
print("✓ Përmbledhja e analizës u ruajt në: analysis_summary.csv")

# Ruajtja e të dhënave të agregatuara
attacks_by_year.to_csv('/content/Sample Data/attacks_by_year.csv', header=['Count'])
attack_types.to_csv('/content/Sample Data/ttack_types.csv', header=['Count'])
countries.to_csv('/content/Sample Data/top_countries.csv', header=['Count'])
industries.to_csv('/content/Sample Data/industries.csv', header=['Count'])
financial_by_attack.to_csv('/content/Sample Data/financial_by_attack.csv')
financial_by_industry.to_csv('/content/Sample Data/financial_by_attack.csv')
financial_by_year.to_csv('/content/Sample Data/financial_by_year.csv')

print("✓ Të gjitha të dhënat e agreguara u ruajtën")

print("\n" + "=" * 80)
print("ANALIZA U PËRFUNDUA ME SUKSES!")
print("=" * 80)
```

## Visualizing Trends

To better understand the trends and distributions of the cyber security threats, we will create several plots:

1.  **Time-series plots** for the number of cyber attacks and financial losses by year.
2.  **Bar plots** for the distribution of attack types.
3.  **Bar plots** for the top 10 most affected countries.
4.  **Bar plots** for targeted industries.
5.  **Bar plots** for financial losses by attack type.
6.  **Bar plots** for financial losses by industry.

**Reasoning**:
First, I will generate time-series plots to visualize the trends of cyber attacks and financial losses over the years, as these were prepared in the previous steps and are foundational to understanding temporal patterns.



```python
print("\n" + "=" * 80)
print("19. VIZUALIZIMI I TENDENCAVE KOHORE")
print("=" * 80)

# Time-series plot for attacks by year
plt.figure(figsize=(12, 6))
sns.lineplot(x=attacks_by_year.index, y=attacks_by_year.values, marker='o', color='skyblue')
plt.title('Numri i Sulmeve Kibernetike Sipas Viteve (2015-2024)', fontsize=16)
plt.xlabel('Viti', fontsize=12)
plt.ylabel('Numri i Sulmeve', fontsize=12)
plt.xticks(attacks_by_year.index, rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Time-series plot for financial losses by year
plt.figure(figsize=(12, 6))
sns.lineplot(x=financial_by_year.index, y=financial_by_year['sum'], marker='o', color='salmon')
plt.title('Humbjet Financiare Totale nga Sulmet Kibernetike Sipas Viteve (2015-2024)', fontsize=16)
plt.xlabel('Viti', fontsize=12)
plt.ylabel('Humbja Financiare Totale (milionë $)', fontsize=12)
plt.xticks(financial_by_year.index, rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("✓ Grafikët e tendencave kohore u gjeneruan.")
```

**Reasoning**:
Now I will generate a bar plot to visualize the distribution of different attack types, as this was specified in the task and will help understand the most common threats.



```python
print("\n" + "=" * 80)
print("20. VIZUALIZIMI I LLOJEVE TË SULMEVE")
print("=" * 80)

# Bar plot for attack types
plt.figure(figsize=(12, 7))
sns.barplot(x=attack_types.index, y=attack_types.values, palette='viridis')
plt.title('Shpërndarja e Llojeve të Sulmeve Kibernetike (2015-2024)', fontsize=16)
plt.xlabel('Lloji i Sulmit', fontsize=12)
plt.ylabel('Numri i Sulmeve', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("✓ Grafikët e llojeve të sulmeve u gjeneruan.")
```
